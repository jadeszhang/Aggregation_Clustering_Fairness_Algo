```python
# import packages
import pandas as pd
```

```python
df = pd.read_csv('UCD_aggregation_all_races_data_input.csv').drop(columns=["Unnamed: 0"])
df = df.rename(columns={"variable": "race"})

df.columns = df.columns.str.lower()
df = df[df['year']!= 2020]
df = df.drop('student_level', axis = 1)
```

```python
# merging the columns we dont need
df_1 = df.groupby(['year','race','first_gen_status',
                'enrollment_status', 'gender', 'residency'], as_index=False)['count'].sum()
```

```python
df_2 = df.groupby(['year','first_gen_status',
                'enrollment_status', 'gender', 'residency'], as_index=False)['count'].sum()
df_2
```

```python
df_1.head()
race_list = df_1['race'].unique()
```

```python
fg_pivot = df_1.pivot_table(index='year', columns=['race','first_gen_status','enrollment_status','gender','residency'], values='count', aggfunc='sum')
fg_pivot['Afghan']['first']['cont']['fem']['int'][2019]
```

```python
fg_pivot.tail(5)
```

```python
df_2 = df_2.drop(['count'], axis = 1)

df_dummy = pd.get_dummies(df_2, columns=['first_gen_status', 'enrollment_status','gender','residency'])
# initialize 

for r in race_list:
    df_dummy[r] = 0

df_dummy
```

```python
# make dummy vars
for (race, fg, enroll, gender, res) in fg_pivot.columns:
    
    # df_dummy[race] = 0 # initialize the col
    # for each year 
    dum_fg = "first_gen_status_" + fg
    dum_enroll = "enrollment_status_" + enroll
    dum_gender = "gender_" + gender
    dum_res = "residency_" + res
    
    for i in fg_pivot.index:
        df_dummy.loc[(df_dummy["year"] == i) & (df_dummy[dum_fg] == 1) & (df_dummy[dum_enroll] == 1) & (df_dummy[dum_gender] == 1) & (df_dummy[dum_res] == 1), race] = fg_pivot[race][fg][enroll][gender][res][i]
```
